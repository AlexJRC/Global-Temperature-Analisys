# Remove Duplicates datas

```python
duplicados = Temperature[Temperature.duplicated(keep='first')]
duplicados
```

```python
Temperature.drop_duplicates(keep='first', inplace=True) 
Temperature
```

# Filter by Brazil Country

```python
pais_brasil= Temperature["Country"]=="Brazil"
Brasil=Temperature[pais_brasil]
Brasil
```

```python
Faltantes= Brasil.isnull().sum()
Faltantes
```

```python
# How all elementes of State are NaN, i will remove this column
del Brasil["State"]
```

```python
AVGM=Brasil["AvgTemperature"] >=70
AVG = Brasil[AVGM]
AVG
```

```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.lineplot(data=AVG["City"], y=AVG["AvgTemperature"], x=AVG["Year"], hue=AVG["City"])
```

```python
#In The Graph, RIO DE JANERIO is the hotter city of brazil (mean). Lets see the mouth more warm.
```

```python
RioJan= AVG["City"]=="Rio de Janeiro"
RJ=AVG[RioJan]
RJ
```

```python
sns.lineplot(data=RJ["Year"], y=RJ["AvgTemperature"], x=RJ["Month"], color="Orange")
```

```python
# The temperature is more larger in first day of year: January in Rio de Janeiro. 
#Lets see the Year more Hotter in month 1, in Rio de Janeiro
```

```python
Janeiro= RJ["Month"]==1
Jan=RJ[Janeiro]
Jan
```

```python
sns.lineplot(data=Jan["City"], y=Jan["AvgTemperature"], x=Jan["Year"], color="orange")
```

```python
# In 2015 was the Year more hotter in RIO.
# What Day in January of 2015 is more hotter
```

```python
Pas=Jan["Year"]==2015
Past=Jan[Pas]
Past
```

```python
sns.lineplot(data=Past["City"], y=Past["AvgTemperature"], x=Past["Day"], color="Tomato")
```
